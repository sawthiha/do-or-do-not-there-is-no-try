```python
import skfuzzy as fuzz
import numpy as np
import matplotlib.pyplot as plot
import matplotlib.patches as patches
```

```python
## Term-sets
# Marks
#----------------------------------------
# Low
# Average
# Excellent
low = np.array([0, 0, 40, 60])
average = np.array([40, 60, 80])
excellent = np.array([60, 80, 100, 100])

# Attendance
#----------------------------------------
# Low
# Average
# Excellent
a_low = np.array([0, 0, 75])
a_average = np.array([70, 75 , 80])
a_excellent = np.array([75, 80, 100, 100])

# Teachers' rating
#----------------------------------------
# NeedImprovement
# Average
# Recommended
r_ni = np.array([0, 0, 40, 45])
r_average = np.array([40, 45, 55, 60])
r_recommended = np.array([55, 70, 100, 100])

y_trapezoid = np.array([0, 1, 1, 0])
y_triangle = np.array([0, 1, 0])
```

```python
fig = plot.figure()
ax = fig.add_subplot(111)
ax.axis([0, 100, 0, 1])
ax.set_xlabel('Marks')
ax.set_ylabel('Membership Function')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

co_low = list(zip(low, y_trapezoid))
co_average = list(zip(average, y_triangle))
co_excellent = list(zip(excellent, y_trapezoid))

patch_low = patches.Polygon(xy = co_low, color = 'red', fill = False, label = 'low', zorder = 0.3)
patch_average = patches.Polygon(xy = co_average, color = 'yellow', fill = False, label = 'average', zorder = 0.4)
patch_excellent = patches.Polygon(xy = co_excellent, color = 'green', fill = False, label = 'excellent', zorder = 0.3)

ax.add_patch(patch_low)
ax.add_patch(patch_average)
ax.add_patch(patch_excellent)

box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

plot.legend(loc = 'upper right', handles=[patch_low, patch_average, patch_excellent], bbox_to_anchor = (1.35, 0.95))
plot.show()
```

```python
fig = plot.figure()
ax = fig.add_subplot(111)
ax.axis([0, 100, 0, 1])
ax.set_xlabel('Attendance')
ax.set_ylabel('Membership Function')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

co_low = list(zip(a_low, y_triangle))
co_average = list(zip(a_average, y_triangle))
co_excellent = list(zip(a_excellent, y_trapezoid))

patch_low = patches.Polygon(xy = co_low, color = 'red', fill = False, label = 'low', zorder = 0.3)
patch_average = patches.Polygon(xy = co_average, color = 'yellow', fill = False, label = 'average', zorder = 0.4)
patch_excellent = patches.Polygon(xy = co_excellent, color = 'green', fill = False, label = 'excellent', zorder = 0.3)

ax.add_patch(patch_low)
ax.add_patch(patch_average)
ax.add_patch(patch_excellent)

box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

plot.legend(loc = 'upper right', handles = [patch_low, patch_average, patch_excellent], bbox_to_anchor = (1.35, 0.95))
plot.show()
```

```python
fig = plot.figure()
ax = fig.add_subplot(111)
ax.axis([0, 100, 0, 1])
ax.set_xlabel('Teachers\' Feedback')
ax.set_ylabel('Membership Function')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

co_low = list(zip(r_ni, y_trapezoid))
co_average = list(zip(r_average, y_trapezoid))
co_excellent = list(zip(r_recommended, y_trapezoid))

patch_low = patches.Polygon(xy = co_low, color = 'red', fill = False, label = 'Need Improvement', zorder = 0.3)
patch_average = patches.Polygon(xy = co_average, color = 'yellow', fill = False, label = 'average', zorder = 0.4)
patch_excellent = patches.Polygon(xy = co_excellent, color = 'green', fill = False, label = 'Recommended', zorder = 0.3)

ax.add_patch(patch_low)
ax.add_patch(patch_average)
ax.add_patch(patch_excellent)

box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

plot.legend(loc = 'upper right', handles = [patch_low, patch_average, patch_excellent], bbox_to_anchor = (1.6, 0.95))
plot.show()
```

```python
m_x = np.arange(0, 1.1, 0.10)
a_x = np.arange(0, 1.1, 0.10)
e_x = np.arange(0, 1.1, 0.10)

p_x = np.arange(0, 1.1, 0.10)

m_low = fuzz.trapmf(m_x, [0, 0, 0.6, 0.75])
m_avg = fuzz.trimf(m_x, [0.6, 0.75, 0.8])
m_exc = fuzz.trapmf(m_x, [0.75, 0.8, 1, 1])

a_low = fuzz.trapmf(a_x, [0, 0, 0.4, 0.6])
a_avg = fuzz.trimf(a_x, [0.4, 0.6, 0.8])
a_exc = fuzz.trapmf(a_x, [0.6, 0.8, 1, 1])

e_low = fuzz.trapmf(e_x, [0, 0, 0.4, 0.6])
e_avg = fuzz.trimf(e_x, [0.4, 0.6, 0.8])
e_exc = fuzz.trapmf(e_x, [0.6, 0.8, 1, 1])
```

```python
fig, (ax0, ax1, ax2, ax3)= plot.subplots(nrows=4, figsize=(8, 9))
ax0.plot(m_x, m_low, 'r', linewidth=1.5, label='low')
ax0.plot(m_x, m_avg, 'y', linewidth=1.5, label='average')
ax0.plot(m_x, m_exc, 'g', linewidth=1.5, label='excellent')
ax0.set_title('Internal Marks')
ax0.legend()

ax1.plot(a_x, a_low, 'r', linewidth=1.5, label='low')
ax1.plot(a_x, a_avg, 'y', linewidth=1.5, label='average')
ax1.plot(a_x, a_exc, 'g', linewidth=1.5, label='excellent')
ax1.set_title('Class Activities')
ax1.legend()

ax2.plot(e_x, e_low, 'r', linewidth=1.5, label='low')
ax2.plot(e_x, e_avg, 'y', linewidth=1.5, label='average')
ax2.plot(e_x, e_exc, 'g', linewidth=1.5, label='excellent')
ax2.set_title('External Marks')
ax2.legend()

ax3.plot(e_x, r_low, 'r', linewidth=1.5, label='low')
ax3.plot(e_x, r_avg, 'y', linewidth=1.5, label='average')
ax3.plot(e_x, r_exc, 'g', linewidth=1.5, label='excellent')
ax3.set_title('Performance')
ax3.legend()

for ax in (ax0, ax1, ax2, ax3):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left

plot.tight_layout()
```
